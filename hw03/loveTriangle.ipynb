## 1 Nonlinear love triangle (5 points)

Please use Euler integration with a small step size (say dt=0.01) and the suggested parameters (a,b,c,d,e,f)=(-3,4,-7,2,2,-1) to simulate Eq.(5) in the following paper:

Sprott, J. C. (2004). Dynamical models of love. Nonlinear dynamics, psychology, and life sciences, 8(3), 303-314.

### 1.1  Coding the simulation (3 points)

Please use differnt colors/shapes to put Rj(t), J, Rg(t), & G(t) curves in the same plot.

### Initial condition 1: ($R_G$, $R_J$, $G$, $J$)  =  (0, 0, -0.01, +0.01)

```python
# Write your codes here
a,b,c,d,e,f=-3,4,-7,2,2,-1  # Model parameters
rg,rj,g,j=[0],[0],[-0.01],[+0.01] # Initial values
dt=.0001 # step size for numerical integration
t=arange(0,50,dt) # integration period
for i in t:
    rg1,rj1, g1,j1=rg[-1],rj[-1],g[-1],j[-1], # the last element

    rg2=rg1+dt*(a*rg1 + b*(g1-j1)*(1-abs(g1-j1)) ) # Euler integration
    rj2=rj1+dt*(a*rj1 + b*(j1-g1)*(1-abs(j1-g1)) ) # Euler integration
    g2=  g1+dt*(e*rg1*(1-abs(rg1)) + f*g1) # Euler integration
    j2=  j1+dt*(c*rj1*(1-abs(rj1)) + d*j1) # Euler integration

    rg.append(rg2)
    rj.append(rj2)
    g.append(g2)
    j.append(j2)

plot(t,rg[0:-1],'bo') # Romeo_Guinevere
plot(t,rj[0:-1],'go') # Romeo_Juliet
plot(t,g[0:-1],'ko') # Guinevere
plot(t,j[0:-1],'ro') # Juliet
legend(['Romeo_Guinevere','Romeo_Juliet','Guinevere','Juliet']);
```

### Plot $G$ vs. $R_G$, $J$ vs. $R_J$

```python
plot(rj[0:-1],j[0:-1],'r-') # Juliet
plot(rg[0:-1],g[0:-1],'b-') # Guinevere
xlim(-3,3)
xlabel('RJ:red;  RG:blue')
ylim(-3,3)
ylabel('J:red;  G:blue')
```

## 2 Fix the simulation of "Growth and underinvestment" (5 points)

Although the simulation and explanation of "underinvested growth" seem reasonable, the simulation doesn't appear to describe "sufficiently invested growth" where the performance varialbe should be able to increase over time. Please do whatever you can to have a model that can simulate both phenomena. 

```python
performance_standard=100 # Model parameter
Capacity,Demand=[80],[1] # Initial values
dt=.1 # step size for numerical integration
t=arange(0,13,dt) # integration period
perf,inves=[],[]
for i in t:
    Capacity1,Demand1=Capacity[-1],Demand[-1] # the last element
    growing_action=0.01*Demand1
    performance=Capacity1-Demand1
    change_rate=0.75*growing_action*performance
    perf.append(performance)
    growth_and_underinverstment=performance_standard-performance
    investment=0.1*growth_and_underinverstment # !!!!!!!! Change 0.01 to 10 doesn't help !!!!!!!!
    inves.append(investment)
    Capacity2=Capacity1+dt*investment # Euler integration
    Demand2=Demand1+dt*change_rate # Euler integration
    Capacity.append(Capacity2)
    Demand.append(Demand2)
plot(t,Capacity[0:-1],'bv')
plot(t,inves,'ro')
plot(t,perf,'g.')
legend(['Capacity','Investment','Performance']);
```

## Write down the equations and observe

#### (1) $\frac{dCapacity}{dt}= Investment$
#### (2) $\frac{dDemand}{dt}= Change \_ rate$
#### (3) $\frac{dChange \_ rate}{dt}= 0.75 * Growing \_ action * Performance$
#### (4) $Growing \_ action = 0.01 * Demand$
#### (5) $Performance = Capacity - Demand$
#### (6) $Investment = 0.1*Growth \_ and \_ Underinvestment$
#### (7) $Growth \_ and \_ Underinvestment = Performance \_ standard-Performance$
<br /><br /><br />

To make the performance increase over time, examine $\frac{dPerformance}{dt}$

$\frac{dPerformance}{dt} = \frac{d(Capacity-Demand)}{dt}$
$= Investment - Change \_ rate $ 

$= Investment - 0.75*0.01*Demand*Performance $

$= 0.1*(Performance \_ standard-Performance) - 0.75*0.01*Demand*Performance $

<br />

It turns out that the Demand inhibits the Performace by ($-0.75*0.01*Demand*Performance$), and since Demand grows very fast (expontial growth, with Capacity the upper bound), the Performace hardly boost with investment by tuning 0.1 to 10 in ($0.1*(Performance \_ standard-Performance)$). 

Hence, investment is modified to follow the Demand
$Investment=(Investment\_Parameter)*(Growth\_and\_Underinverstment + Performance\_standard*Demand)$

The modified $\frac{dPerformance}{dt}$ is 
$= (Investment\_Parameter)*(Performance \_ standard*(1+Demand)-Performance) - 0.75*0.01*Demand*Performance $

```python
# Write your working simulation here
performance_standard=100 # Model parameter
Capacity,Demand=[80],[1] # Initial values
dt=.001 # step size for numerical integration
t=arange(0,13,dt) # integration period
perf,inves=[],[]
for i in t:
    Capacity1,Demand1=Capacity[-1],Demand[-1] # the last element
    growing_action=0.01*Demand1
    performance=Capacity1-Demand1
    change_rate=0.75*growing_action*performance
    perf.append(performance)
    growth_and_underinverstment=performance_standard-performance
    investment=0.008*(growth_and_underinverstment + performance_standard*Demand1)
    inves.append(investment)
    Capacity2=Capacity1+dt*investment# Euler integration
    Demand2=Demand1+dt*change_rate # Euler integration
    Capacity.append(Capacity2)
    Demand.append(Demand2)
plot(t,Capacity[0:-1],'bv')
plot(t,inves,'ro')
plot(t,perf,'g.')
plot(t,Demand[0:-1],'c.')
legend(['Capacity','Investment','Performance','Demand']);
```

### Vary the investment parameter to simulate the suffiecient-investment and under-investment cases
$Investment\_Parameter$ $(I\_P)$ = 0.001 (under-investment) ~ 0.011 (sufficeint investment)

```python
def Investment_model(investment_parameter):
    global Capacity, Demand, inves, perf
    performance_standard=100 # Model parameter
    Capacity,Demand=[80],[1] # Initial values
    dt=.0001 # step size for numerical integration
    t=arange(0,13,dt) # integration period
    perf,inves=[],[]
    for i in t:
        Capacity1,Demand1=Capacity[-1],Demand[-1] # the last element
        growing_action=0.01*Demand1
        performance=Capacity1-Demand1
        change_rate=0.75*growing_action*performance
        perf.append(performance)
        growth_and_underinverstment=performance_standard-performance
        investment=investment_parameter*(growth_and_underinverstment + performance_standard*Demand1)
        inves.append(investment)
        Capacity2=Capacity1+dt*investment# Euler integration
        Demand2=Demand1+dt*change_rate # Euler integration
        Capacity.append(Capacity2)
        Demand.append(Demand2)
    
    return Capacity, Demand, inves, perf
```

```python
dt = .0001
t = arange(0,13,dt)
Capacity, Demand, inves, perf = Investment_model(0.001)
plot(t,perf)
Capacity, Demand, inves, perf = Investment_model(0.003)
plot(t,perf)
Capacity, Demand, inves, perf = Investment_model(0.005)
plot(t,perf)
Capacity, Demand, inves, perf = Investment_model(0.007)
plot(t,perf)
Capacity, Demand, inves, perf = Investment_model(0.009)
plot(t,perf)
Capacity, Demand, inves, perf = Investment_model(0.011)
plot(t,perf)
#plot(t,Demand[0:-1],'c.')
legend(['I_P = 0.001', 'I_P = 0.003', 'I_P = 0.005', 'I_P = 0.007', 'I_P = 0.009', 'I_P = 0.011']);
```

All the performance saturates at certain level, which can be calculated by setting $\frac{dPerformance}{dt} = 0$
<br />
$Performance_{\frac{dPerformance}{dt} = 0} = \frac{Performance\_standard*(1+Demand)*Investment\_Parameter}{0.75*0.01*Demand + Investment\_Parameter}$
<br /><br />
As $Demand$ grows fast enough, 
$Performance \rightarrow \frac{Performance\_standard*(Demand)*Investment\_Parameter}{0.75*0.01*Demand} = \frac{Performance\_standard**Investment\_Parameter}{0.75*0.01}$


```python
investment_parameter = [0.001, 0.003, 0.005, 0.007, 0.009, 0.011]
performance_saturate = [round(i*performance_standard/0.75/0.01,2) for i in investment_parameter]
print(performance_saturate)
```
